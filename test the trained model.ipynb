```python
#First testing phase with the trained model

import os
import numpy as np
from PIL import Image
import joblib  # Use joblib to load the trained model

# Define the path to the directory containing new images
new_images_dir = "new_images/"

# List all the filenames in the new images directory
new_filenames = [filename for filename in os.listdir(new_images_dir) if filename.endswith(".jpg")]

# Load the trained model
loaded_model = joblib.load("trained_img_model.pkl")  # Replace with the actual model file path

# Initialize a list to store image paths and predictions
predictions = []

# Define a function for preprocessing new images
def preprocess_new_images(data_filenames):
    images = []
    for filename in data_filenames:
        try:
            img = Image.open(os.path.join(new_images_dir, filename))
            #img = img.resize((64, 64))  # Resize images to the same dimensions as the training data
            img = np.array(img)
            images.append(img)
        except Exception as e:
            print(f"Error processing {filename}: {str(e)}")
            continue
    return np.array(images)

# Preprocess the new images
new_images = preprocess_new_images(new_filenames)

# Make predictions on the new images
new_image_predictions = loaded_model.predict(new_images.reshape(new_images.shape[0], -1))

# Print the predictions
for filename, prediction in zip(new_filenames, new_image_predictions):
    if prediction == 1:
        print(f"{filename}: Contains motion")
    else:
        print(f"{filename}: Does not contain motion")
```
